Exercise 3

```python
def count(line):

    word_list = line.split()


    words = {}


    for word in word_list:

        words[word] = words.get(word, 0) + 1

    return words

line = input("Enter a line of text: ")


words = count(line)

print("Word counts:")
for word, count in words.items():
    print(f"{word}: {count}")
```
