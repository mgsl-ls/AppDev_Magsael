```python
def fr_name(full_name):

    f_name = full_name.strip().capitalize()


    space = f_name.find(' ')
    if space != -1:
        f_name = f_name[:space + 1] + f_name[space + 1:].lower()

    return f_name

def main():
    while True:
        full_name = input("Enter a full name (or 'quit' to exit): ")
        if full_name.lower() == 'quit':
            break
        f_name = fr_name(full_name)
        print("Fixed name:", f_name)

if __name__ == "__main__":
    main()
```
