What are the constants in the math module?

What's does math.exp do? What about math.pow?
- math.exp returns e^x in which e is an Euler's number and x is an argument that can be passed to a function.
How do you calculate logarithms? Logarithms base 10? Natural Logarithms?
- In calculating the natural logarithm we can use math.log while in calculating the logarithm base 10 we can use the math.log10

Example of calculating logarithm is shown below

```python
import math


x = 100


n_l = math.log(x)
print("Natural Logarithm of", x, ":", n_l)


log10 = math.log10(x)
print("Logarithm base 10 of", x, ":", log10)


base = 5
log2 = math.log(x) / math.log(base)
print("Logarithm base", base, "of", x, ":", log2)
```
