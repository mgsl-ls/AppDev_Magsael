```python
x = int(input('Enter value of x: '))

if x > 0 and x < 10 :
    print('Above 0')
elif x > 10 and x < 20 :
    print('Above 10')
elif x > 20 and x < 50 :
    print('Above 20')
elif x > 50 :
    print('Above 50')
else :
    print('Something else')
```

Exercise 2

```python
hour = float(input('Enter hour: '))
rate = float(input('Enter rate: '))

if hour > 40 :
  regular = 40 * rate
  over = (hour - 40) * (rate * 1.5)
  pay = regular + over
else :
  pay = hour * rate

print(f'Pay:{pay:.2f}')

```
