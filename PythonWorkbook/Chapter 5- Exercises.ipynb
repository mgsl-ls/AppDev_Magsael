The difference between "is" and == is that "is" is being used in checking for identity in a sense that it will check if the variable points at the same object in the memory. While "==" is being used in checking if the value of the two variables are equal in other words it checks for equality.

When using "is" for example a is c as mentioned above it checks for identity and since c = a and a has the value 2 its endpoint is in the same integer which is the 2.

When using "==" since as mentioned above we are dealing with equality the given example variables a and b contains equal value which is 2 regardless of its data type since we are only talking about equality of the values.
In the case of a = c the value that is stored inside c is a which makes them hold the same value that makes them equal.

```python
from string import digits
total = 0.0
count = 0
while(True):
    inputNum =(input("Enter a number: "))


    if(inputNum == 'done'):
      break
    elif inputNum.isdigit():
       number = float(inputNum)
       total = total + number
       count += 1
    else:
       print("bad data")

if(total != 0):
      print(f"Total: {total} Count: {int(count)} Average: {total/count}")
```

```python
number = []
count = 0
max = None
min = None
while(True):
  inputNum = input("Enter number: ")

  if(inputNum == 'done'):
    break
  elif inputNum.isdigit():
     number += [float(inputNum)]
     count += 1
  else:
    print("bad data")

if(count != 0):
  for x in number:
    if min is None:
      min = x
    elif x < min:
      min = x
    if max is None:
      max = x
    elif x > max:
      max = x
print(f"Count: {count} Maximum: {max} Minimum: {min}")
```
