```python
def Computepay(hour, rate):
  if hour > 40 :
    regular = 40 * rate
    over = (hour - 40) * (rate * 1.5)
    pay = regular + over
  else :
    pay = hour * rate

  return pay

h = float(input("Enter hours: "))
r = float(input("Enter rate: "))

print(Computepay(h, r))




```

```python
def computegrade(score):

  if score > 0 and score <= 74:
    grade = '3.0'
  elif score > 74 and score <= 77:
    grade = '2.5'
  elif score > 77 and score <= 81:
    grade = '2.25'
  elif score > 81 and score <= 85:
    grade = '2.0'
  elif score > 85 and score <= 89:
    grade = '1.75'
  elif score > 89 and score <= 93:
    grade = '1.5'
  elif score > 93 and score <= 97:
    grade = '1.25'
  elif score > 97 and score <= 100:
    grade = '1'

  return grade

score = int(input('Enter score between 0 and 100: '))
if score > 100 and score < 0:
  print("Error")

else:
  print(computegrade(score))
```
